# Phase 1 & 2: Data Loading and EDA
In this notebook, we load the `customer_details.csv` dataset and perform initial Exploratory Data Analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_theme(style="whitegrid")

# Load the dataset
data_path = '../data/raw/customer_details.csv'
df = pd.read_csv(data_path)
display(df.head())
```

```python
# Check data types and missing values
df.info()
```

```python
# Summary statistics for numerical variables
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

plt.figure(figsize=(15, 5))

# Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Customer Age')

# Gender Distribution
plt.subplot(1, 2, 2)
sns.countplot(x='Gender', data=df, hue='Gender', palette='pastel', legend=False)
plt.title('Customer Gender Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

# Purchase Amount Distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Purchase Amount (USD)'], bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Purchase Amount (USD)')

# Purchases by Category
plt.subplot(1, 2, 2)
sns.countplot(y='Category', data=df, hue='Category', palette='muted', legend=False, order=df['Category'].value_counts().index)
plt.title('Purchases by Category')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

# Spending by Gender
plt.subplot(1, 2, 1)
sns.boxplot(x='Gender', y='Purchase Amount (USD)', data=df, hue='Gender', palette='pastel', legend=False)
plt.title('Purchase Amount by Gender')

# Spending by Category
plt.subplot(1, 2, 2)
sns.boxplot(x='Purchase Amount (USD)', y='Category', data=df, hue='Category', palette='muted', legend=False)
plt.title('Purchase Amount by Category')

plt.tight_layout()
plt.show()
```
